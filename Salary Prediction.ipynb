### Salary Prediction using Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv")
```

```python
# linear regression

X = df.Level.values[:,np.newaxis]
Y = df.Salary.values

lin_reg.fit(X, Y)

pred_val = lin_reg.predict(X)
plt.scatter(X, Y, color='red')
plt.plot(X, pred_val, color='green', label='Regression line')
plt.title('Simple linear Regression')
plt.xlabel('Level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
# Polynomial Regression: degree 2
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)

Xpoly2 = poly_reg.fit_transform(X)

lin_reg.fit(Xpoly2,Y)

pred_val2 = lin_reg.predict(Xpoly2)
plt.scatter(X, Y, color='red')
plt.plot(X, pred_val2, color='green', label='Regression line')
plt.title('Plynomial Regression')
plt.xlabel('Level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
# Polynomial Regression: degree 3
poly_reg3 = PolynomialFeatures(degree=3)

Xpoly3 = poly_reg3.fit_transform(X)

lin_reg.fit(Xpoly3, Y)

pred_val3 = lin_reg.predict(Xpoly3)
plt.scatter(X, Y, color='red')
plt.plot(X, pred_val3, color='green', label='Regression line')
plt.title('Plynomial Regression')
plt.xlabel('Level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
# Polynomial Regression: degree 4
poly_reg4 = PolynomialFeatures(degree=4)

Xpoly4 = poly_reg4.fit_transform(X)

lin_reg.fit(Xpoly4,Y)

pred_val4 = lin_reg.predict(Xpoly4)
plt.scatter(X, Y, color='red')
plt.plot(X, pred_val4, color='green', label='Regression line')
plt.title('Plynomial Regression')
plt.xlabel('Level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
# Comparing models through accuracy
from sklearn.metrics import r2_score

print('Accuracy Of Linear Regression Model : ', r2_score(Y, pred_val))
print('Accuracy Of Degree 2 Polynomial Regression Model : ', r2_score(Y, pred_val2))
print('Accuracy Of Degree 3 Polynomial Regression Model : ', r2_score(Y, pred_val3))
print('Accuracy Of Degree 4 Polynomial Regression Model : ', r2_score(Y, pred_val4))
```

```python
# Using Degree 4 Polynomial Regreesion Model for predictions

lvl = int(input('Enter the level Number: '))
prediction = lin_reg.predict(poly_reg4.fit_transform([[lvl]]))
print('For level: ', lvl, ', salary will be: ', int(prediction[0]))
```
