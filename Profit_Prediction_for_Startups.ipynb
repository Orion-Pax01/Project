### Profit Prediction for Startups using Multiple Linear Regression

```python
#Importing the libraries
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Startups_Data.csv')
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
x = ['R&D Spend', 'Administration', 'Marketing Spend', 'State']
X = df[x].values
```

```python
Y = df.Profit.values
```

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
from sklearn.compose import ColumnTransformer
label_encoder = LabelEncoder()

X[:,3] = label_encoder.fit_transform(X[:,3])

transformer = ColumnTransformer(transformers=[("OneHot", OneHotEncoder(), [3])], remainder='passthrough')

X = transformer.fit_transform(X)
X = X.astype(int)
```

```python
# Splittin the dataset
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# Building the model
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(X_train,Y_train)
```

```python
pred_val = mlr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(Y_test, pred_val)
print('Accuracy of model :', accuracy*100)
```
